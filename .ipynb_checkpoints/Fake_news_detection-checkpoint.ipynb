# Fake news Detection

### Importing required library
Here I am going to importing some of the required library, if extra library is required to install It will be install later on.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
from joblib import load, dump
```

### Inserting fake and real dataset

```python
df_fake = pd.read_csv("./data/Fake.csv")
df_true = pd.read_csv("./data/True.csv")
```

```python
df_fake.head(5)
```

```python
df_true.head(5)
```

Inserting a column called "class" for fake and real news dataset to categories fake and true news. 

```python
df_fake["class"] = 0
df_true["class"] = 1
```

Removing last 10 rows from both the dataset, for manual testing  

```python
df_fake.shape, df_true.shape
```

```python
df_fake_manual_testing = df_fake.tail(10)
for i in range(23480,23470,-1):
    df_fake.drop([i], axis = 0, inplace = True)
df_true_manual_testing = df_true.tail(10)
for i in range(21416,21406,-1):
    df_true.drop([i], axis = 0, inplace = True)
```

```python
df_fake.shape, df_true.shape
```

Merging the manual testing dataframe in single dataset and save it in a csv file

```python
df_fake_manual_testing["class"] = 0
df_true_manual_testing["class"] = 1;
```

```python
df_fake_manual_testing.head(10)
```

```python
df_true_manual_testing.head(10)
```

```python
df_manual_testing = pd.concat([df_fake_manual_testing,df_true_manual_testing], axis = 0)
df_manual_testing.to_csv("./data/manual_testing.csv")
```

Merging the main fake and true dataframe

```python
df_marge = pd.concat([df_fake, df_true], axis =0 )
df_marge.head(10)
```

```python
df_marge.columns
```

#### "title",  "subject" and "date" columns is not required for detecting the fake news, so I am going to drop the columns.

```python
df = df_marge.drop(["title", "subject","date"], axis = 1)
```

```python
df.isnull().sum()
```

#### Randomly shuffling the dataframe 

```python
df = df.sample(frac = 1)
```

```python
df.head()
```

```python
df.reset_index(inplace = True)
df.drop(["index"], axis = 1, inplace = True)
```

```python
df.columns
```

```python
df.head()
```

#### Creating a function to convert the text in lowercase, remove the extra space, special chr., ulr and links.

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)    
    return text
```

```python
df["text"] = df["text"].apply(wordopt)
```

#### Defining dependent and independent variable as x and y

```python
x = df["text"]
y = df["class"]
```

#### Splitting the dataset into training set and testing set. 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25,random_state=0)
```

#### Convert text to vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
dump(vectorization, filename="./models/tfid.joblib")
```

### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR_score= LR.score(xv_test, y_test)
LR_score
```

```python
print(classification_report(y_test, pred_lr))
```

```python
dump(LR, filename="./models/LinearRegression.joblib")
```

### 2. Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT_score= DT.score(xv_test, y_test)
DT_score
```

```python
print(classification_report(y_test, pred_dt))
```

```python
dump(DT, filename="./models/DecisionTreeClassifier.joblib")
```

### 3. Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(xv_train, y_train)
```

```python
pred_gbc = GBC.predict(xv_test)
```

```python
GBC_score= GBC.score(xv_test, y_test)
GBC_score
```

```python
print(classification_report(y_test, pred_gbc))
```

```python
dump(GBC, filename="./models/GradientBoostingClassifier.joblib")
```

### 4. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train, y_train)
```

```python
pred_rfc = RFC.predict(xv_test)
```

```python
RFC_score=RFC.score(xv_test, y_test)
RFC_score
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
dump(GBC, filename="./models/RandomForestClassifier.joblib")
```

```python
scores_rec={"Logistic Regression":[round(LR_score*100,2)],
           "Decision Tree":[round(DT_score*100,2)],
           "Greadent Boost":[round(GBC_score*100,2)],
           "Random Forest":[round(RFC_score*100,2)]}

df_score=pd.DataFrame(scores_rec)
df_score.to_csv("./data/df_score.csv",index=False)
```

```python
# Importing neccessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import re
import string
from joblib import load

# Creating function to remove abreviation
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)    
    return text

# Creating function for output
def output_lable(n):
    if n == 0:
        return "\033[91m"+'\033[1m'+"Fake News"+'\033[0m'
    elif n == 1:
        return "\033[92m"+'\033[1m'+"Real News"+'\033[0m'

def load_function():
  ''' This function loads the saved joblibs model from the models directory
      and returns them as an output.
  '''
  loaded_LR=load(filename="./models/LinearRegression.joblib")
  loaded_DT=load(filename="./models/DecisionTreeClassifier.joblib")
  loaded_GBC=load(filename="./models/GradientBoostingClassifier.joblib")
  loaded_RFC=load(filename="./models/RandomForestClassifier.joblib")
  return loaded_LR,loaded_DT,loaded_GBC,loaded_RFC
    
# Creating Manual testing function
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt) 
    new_x_test = new_def_test["text"]
    vectorization = load(filename="./models/tfid.joblib")
    new_xv_test = vectorization.transform(new_x_test)
    
    #Loading saved models
    # loaded_LR=load(filename="./models/LinearRegression.joblib")
    # loaded_DT=load(filename="./models/DecisionTreeClassifier.joblib")
    # loaded_GBC=load(filename="./models/GradientBoostingClassifier.joblib")
    # loaded_RFC=load(filename="./models/RandomForestClassifier.joblib")
    loaded_LR,loaded_DT,loaded_GBC,loaded_RFC= load_function()
    
    # predicting the result
    pred_LR = loaded_LR.predict(new_xv_test)
    pred_DT = loaded_DT.predict(new_xv_test)
    pred_GBC = loaded_GBC.predict(new_xv_test)
    pred_RFC = loaded_RFC.predict(new_xv_test)
    
    #predicting the probability of the result
    DT_proba=loaded_DT.predict_proba(new_xv_test)[0][1]
    LR_proba=loaded_LR.predict_proba(new_xv_test)[0][1]
    GBC_proba=loaded_GBC.predict_proba(new_xv_test)[0][1]
    RFC_proba=loaded_RFC.predict_proba(new_xv_test)[0][1]
    
    #Creating values for pie chart
    prob_percent=round(((DT_proba+LR_proba+GBC_proba+RFC_proba)/4)*100,2)
    exp_val=[prob_percent,100-prob_percent]
    exp_labels=["Real","Fake"]
    
    #ploting pie chart
    plt.figure(figsize=(7,7))
    myexplode = [0.2, 0]
    plt.pie(exp_val,labels=exp_labels, explode=myexplode,shadow=True)
    plt.legend(["Real", "Fake"], loc='upper left')
    txt=f"The Probability of the news being Real according to the system is {prob_percent}%\nIf it's more than 50% it could be consider as a Real News"
    plt.text(0,-1.21,txt,horizontalalignment='center')
    plt.title("Fake News Detection")
    
    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_LR[0]), 
                                                                                                              output_lable(pred_DT[0]), 
                                                                                                              output_lable(pred_GBC[0]), 
                                                                                                              output_lable(pred_RFC[0])))
```
